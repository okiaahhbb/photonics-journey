```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

plt.plot(x, y)
plt.title("Simple sine wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 1000)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.plot(x, y3, label="sin(2x)")
plt.title("Wave superposition")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# দুটো light wave যখন একসাথে মেলে — interference
x = np.linspace(0, 4 * np.pi, 1000)

wave1 = np.sin(x)
wave2 = -np.sin(x )  # সামান্য phase shift

combined = wave1 + wave2  # siএটাই interference

plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x, wave1, color='blue')
plt.title("Wave 1")
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x, wave2, color='orange')
plt.title("Wave 2 (phase shifted)")
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(x, combined, color='green')
plt.title("Combined — Interference")
plt.grid(True)

plt.tight_layout()
plt.show()
```
