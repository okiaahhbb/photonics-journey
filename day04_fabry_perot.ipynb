```python
import numpy as np
import matplotlib.pyplot as plt

# আয়নার reflectivity
R = 0.9      # 90% reflective
F = 4 * R / (1 - R)**2    # Finesse coefficient

# wavelength range
wavelength = np.linspace(500e-9, 700e-9, 10000)  # 500nm to 700nm

# cavity length — দুই আয়নার দূরত্ব
L = 1e-3     # 1 mm

# phase difference
delta = (4 * np.pi * L) / wavelength

# Transmission
T = 1 / (1 + F * np.sin(delta / 2)**2)

plt.figure(figsize=(10, 5))
plt.plot(wavelength * 1e9, T, color='blue')
plt.title("Fabry-Pérot Transmission")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Transmission (0 to 1)")
plt.grid(True)
plt.show()
```
