```python
from LightPipes import *
import matplotlib.pyplot as plt
import numpy as np

# Setup
wavelength = 632e-9      # লাল laser
size = 10e-3             # 10mm grid
N = 256               # resolution

# Beam তৈরি করো
F = Begin(size, wavelength, N)
F = GaussBeam(F, 2e-3)   # 2mm Gaussian beam

# Plot
plt.figure(figsize=(5, 5))
plt.imshow(Intensity(F), cmap='hot')
plt.title("Gaussian Beam — Before Lens")
plt.colorbar()
plt.show()
```

```python
# Lens দাও
f = 50e-3    # 50mm focal length
F = Lens(F, f)

# Focal point-এ propagate করো
F = Forvard(F, f)

# Plot
plt.figure(figsize=(5, 5))
plt.imshow(Intensity(F), cmap='hot')
plt.title("After Lens — Focused Beam")
plt.colorbar()
plt.show()
```

```python
F = Begin(size, wavelength, N)
F = GaussBeam(F, 2e-3)
F = Lens(F, f)

distances = np.linspace(0, 100e-3, 50)
beam_sizes = []

for d in distances:
    F_temp = Begin(size, wavelength, N)
    F_temp = GaussBeam(F_temp, 2e-3)
    F_temp = Lens(F_temp, f)
    F_temp = Forvard(F_temp, d)
    I = Intensity(F_temp)
    beam_sizes.append(I.max())

plt.figure(figsize=(10, 4))
plt.plot(distances * 1000, beam_sizes, color='red')
plt.axvline(x=50, color='blue', linestyle='--', label='Focal point = 50mm')
plt.title("Beam Intensity vs Distance")
plt.xlabel("Distance (mm)")
plt.ylabel("Peak Intensity")
plt.legend()
plt.grid(True)
plt.show()
```
