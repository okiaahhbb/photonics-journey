```python
import numpy as np
import matplotlib.pyplot as plt

# EDFA Gain Spectrum
# Erbium atom কোন wavelength বেশি amplify করে

wavelength = np.linspace(1520, 1570, 1000)

# EDFA gain profile — real experimental data approximate
gain = (20 * np.exp(-((wavelength - 1532)**2) / (8**2)) + 
        25 * np.exp(-((wavelength - 1550)**2) / (12**2)))

plt.figure(figsize=(10, 5))
plt.plot(wavelength, gain, 'r-', linewidth=2)
plt.axvline(x=1550, color='blue', linestyle='--', 
            label='1550nm — internet wavelength')
plt.title("EDFA Gain Spectrum")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Gain (dB)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gain Flattening
target_gain = 20  # সব channel এই gain পাবে

# GFF filter
gff = target_gain - gain

plt.figure(figsize=(10, 5))
plt.plot(wavelength, gain, 'r-', linewidth=2, label='EDFA gain')
plt.plot(wavelength, gff, 'b-', linewidth=2, label='GFF filter')
plt.plot(wavelength, gain + gff, 'g-', 
         linewidth=2, label='After GFF (flat)')
plt.axhline(y=target_gain, color='gray', 
            linestyle='--', alpha=0.5)
plt.title("Gain Flattening Filter")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Gain (dB)")
plt.legend()
plt.grid(True)
plt.show()
```
