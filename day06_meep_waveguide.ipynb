```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt

print("MEEP version:", mp.__version__)
```

```python
# Waveguide simulation
cell = mp.Vector3(16, 8, 0)

geometry = [
    mp.Block(
        mp.Vector3(mp.inf, 1, mp.inf),
        center=mp.Vector3(),
        material=mp.Medium(epsilon=12)
    )
]

sources = [
    mp.Source(
        mp.ContinuousSource(frequency=0.15),
        component=mp.Ez,
        center=mp.Vector3(-7, 0)
    )
]

pml_layers = [mp.PML(1.0)]

resolution = 10

sim = mp.Simulation(
    cell_size=cell,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution
)

sim.run(until=200)

plt.figure(figsize=(10, 6))
sim.plot2D(fields=mp.Ez)
plt.title("Waveguide — Light propagation")
plt.show()
```

```python
# Waveguide চওড়া করো
cell = mp.Vector3(16, 8, 0)

geometry = [
    mp.Block(
        mp.Vector3(mp.inf, 2, mp.inf),  # 1 এর বদলে 2
        center=mp.Vector3(),
        material=mp.Medium(epsilon=12)
    )
]

sources = [
    mp.Source(
        mp.ContinuousSource(frequency=0.15),
        component=mp.Ez,
        center=mp.Vector3(-7, 0)
    )
]

pml_layers = [mp.PML(1.0)]
resolution = 10

sim = mp.Simulation(
    cell_size=cell,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution
)

sim.run(until=200)

plt.figure(figsize=(10, 6))
sim.plot2D(fields=mp.Ez)
plt.title("Wider Waveguide")
plt.show()
```
