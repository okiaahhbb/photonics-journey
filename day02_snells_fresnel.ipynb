```python
import numpy as np
import matplotlib.pyplot as plt

# Refractive indices
n1 = 1.0   # বাতাস
n2 = 1.5   # কাচ

# আসার angle — 0 থেকে 90 degree
theta1 = np.linspace(0, np.pi/2, 100)

# Snell's Law দিয়ে বাঁকার angle বের করো
sin_theta2 = (n1 / n2) * np.sin(theta1)
theta2 = np.arcsin(sin_theta2)

# Degree-এ convert করো
theta1_deg = np.degrees(theta1)
theta2_deg = np.degrees(theta2)

plt.plot(theta1_deg, theta2_deg, color='blue')
plt.title("Snell's Law — air to glass")
plt.xlabel("Incident angle (degrees)")
plt.ylabel("Refracted angle (degrees)")
plt.grid(True)
plt.show()
```

```python
print('Total Internal Reflection')
import numpy as np
import matplotlib.pyplot as plt

n1 = 1.5  # কাচ
n2 = 1.0  # বাতাস

# Critical angle বের করো
critical_angle = np.degrees(np.arcsin(n2 / n1))

# 0 থেকে critical angle পর্যন্ত
theta1 = np.linspace(0, np.radians(critical_angle), 100)
sin_theta2 = (n1 / n2) * np.sin(theta1)
theta2 = np.degrees(np.arcsin(np.clip(sin_theta2, -1, 1)))
theta1_deg = np.degrees(theta1)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(theta1_deg, theta2, color='blue')
plt.axvline(x=critical_angle, color='red', linestyle='--', label=f'Critical angle = {critical_angle:.1f}°')
plt.title("Total Internal Reflection")
plt.xlabel("Incident angle (degrees)")
plt.ylabel("Refracted angle (degrees)")


plt.subplot(1, 2, 2)
theta_range = np.linspace(0, 90, 500)
reflected = np.where(theta_range >= critical_angle, 100, 0)
plt.plot(theta_range, reflected, color='green')
plt.axvline(x=critical_angle, color='red', linestyle='--')
plt.title("hjg")
plt.xlabel("Incident angle (degrees)")
plt.ylabel("100 = ")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Refractive indices
n1 = 1.0  # air
n2 = 1.5  # glass

theta1 = np.linspace(0, np.pi/2, 500)
cos1=np.cos(theta1)
cos2=np.sqrt(1-(n1/n2 *np.sin(theta1))**2)

# s-polarization (আলোর এক দিক)
Rs = ((n1*cos1 - n2*cos2) / (n1*cos1 + n2*cos2))**2

# p-polarization (আলোর আরেক দিক)
Rp = ((n2*cos1 - n1*cos2) / (n2*cos1 + n1*cos2))**2

theta1_deg = np.degrees(theta1)

plt.figure(figsize=(8, 5))
plt.plot(theta1_deg, Rs, color='blue', label='Rs (s-polarization)')
plt.plot(theta1_deg, Rp, color='red', label='Rp (p-polarization)')
plt.title("Fresnel Equations — Reflection")
plt.xlabel("Incident angle (degrees)")
plt.ylabel("Reflectance (0 to 1)")
plt.legend()
plt.grid(True)
plt.show()
```
