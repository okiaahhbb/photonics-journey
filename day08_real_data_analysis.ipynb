```python
import numpy as np
import matplotlib.pyplot as plt

# Glass এর refractive index vs wavelength
# এটা real experimental data

wavelength = np.array([400, 450, 500, 550, 600, 650, 700, 750, 800])
n_glass = np.array([1.530, 1.522, 1.516, 1.511, 1.508, 1.505, 1.503, 1.501, 1.500])

plt.figure(figsize=(8, 5))
plt.plot(wavelength, n_glass, 'bo-', label='Glass (BK7)')
plt.title("Refractive Index vs Wavelength")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Refractive Index (n)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# তিনটা material একসাথে compare করো
n_water = np.array([1.343, 1.338, 1.335, 1.333, 1.332, 1.331, 1.330, 1.329, 1.329])
n_diamond = np.array([2.465, 2.441, 2.426, 2.416, 2.409, 2.404, 2.400, 2.397, 2.394])

plt.figure(figsize=(8, 5))
plt.plot(wavelength, n_glass, 'bo-', label='Glass (BK7)')
plt.plot(wavelength, n_water, 'rs-', label='Water')
plt.plot(wavelength, n_diamond, 'g^-', label='Diamond')
plt.title("Dispersion — Different Materials")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Refractive Index (n)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Real data দিয়ে Snell's Law
# আলো বাতাস থেকে glass-এ যাচ্ছে
# incident angle = 45 degrees

theta_i = 45  # degrees
theta_i_rad = np.radians(theta_i)
n_air = 1.0

# প্রতিটা wavelength-এ refraction angle বের করো
theta_r = np.degrees(np.arcsin(n_air * np.sin(theta_i_rad) / n_glass))

plt.figure(figsize=(8, 5))
plt.plot(wavelength, theta_r, 'bo-')
plt.title("Refraction Angle vs Wavelength (45° incident)")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Refraction angle (degrees)")
plt.grid(True)
plt.show()
```
