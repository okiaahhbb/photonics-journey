```python
import numpy as np
import matplotlib.pyplot as plt

# Laser parameters
wavelength = 632e-9   # 632 nm — লাল laser
w0 = 0.1e-3             # 1 mm beam waist

# Rayleigh range
zR = np.pi * w0**2 / wavelength
print(f"Rayleigh range: {zR:.4f} meters")
```

```python
import numpy as np
import matplotlib.pyplot as plt

z=np.linspace(-0.20,0.20,1000)
w=w0*np.sqrt(1+(z/zR)**2)

#plot
plt.figure(figsize=(10,5))
plt.plot(z,w*1000,color='red',label='Beam radius (mm)')
plt.plot(z,-w*1000,color='red',label='Beam radius (mm)')
plt.axvline(x=0, color='green', linestyle='--', label='Beam waist')
plt.axvline(x=zR, color='blue', linestyle='--', label=f'Rayleigh range = {zR:.1f} m')
plt.axvline(x=-zR, color='blue', linestyle='--')
plt.fill_between(z, w * 1000, -w * 1000, alpha=0.2, color='red')
plt.title("Gaussian Beam Profile")
plt.xlabel("Distance z (meters)")
plt.ylabel("Beam radius (mm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# তিনটা আলাদা beam একসাথে compare করো
wavelength = 632e-9

beams = [0.5e-3, 1e-3, 2e-3]
colors = ['red', 'blue', 'green']
labels = ['w0 = 0.5mm', 'w0 = 1mm', 'w0 = 2mm']

z = np.linspace(-20, 20, 1000)

plt.figure(figsize=(10, 5))
for w0, color, label in zip(beams, colors, labels):
    zR = np.pi * w0**2 / wavelength
    w = w0 * np.sqrt(1 + (z / zR)**2)
    plt.plot(z, w * 1000, color=color, label=label)

plt.title("Gaussian Beam — Different Beam Waists")
plt.xlabel("Distance z (meters)")
plt.ylabel("Beam radius (mm)")
plt.legend()
plt.grid(True)
plt.show()
```
